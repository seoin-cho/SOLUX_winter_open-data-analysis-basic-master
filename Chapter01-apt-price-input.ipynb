## 데이터 타입 변경
#### 분양가격이 object(문자) 타입으로 되어있다. object 타입은 계산이 불가하므로 수치 타입으로 변경한다. 단, 결측치가 섞여있을 때는 astype(int)로 변환이 제대로 되지 않음. 따라서 pd.to_numeric을 통해 타이터 타입 변경

```python
# NaN의 테이터 타입이 float
type(pd.np.nan)
```

```python
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
df_last["분양가격"] # 새로운 column에 담아줌

# shift+tab을 통해 옵션을 볼 수 있다.
# 'coerce'는 에러를 무시하고 강제로 값을 변환하겠다.
```

## 평당 분양가격 구하기
#### 공공데이터포털에 올라와 있는 2013년부터의 데이터는 평당분양가격 기준으로 되어 있습니다. 분양가격을 평당기준으로 보기위해 3.3을 곱해서 "평당분양가격" 컬럼을 만들어 추가해 줍니다.

```python
df_last["평당분양가격"] = df_last["분양가격"] * 3.3
df_last.head(1)
```

## 규모구분을 전용면적 column으로 변경

```python
df_last["규모구분"].unique()
```

```python
# '전용면적' ,'초과', '이하'라는 단어 제거
# 변경된 데이터를 특정 칼럼에 적용해야만 변경된 값이 적용됨.
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ", "").str.strip()  #공백, 앞뒤 공백 제거
df_last["전용면적"]
```

## groupby로 데이터 집계하기

```python
# 지역명으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다.
# df.groupby(["인덱스로 사용할 컬럼명"])["계산할 컬럼 값"].연산()
df_last.groupby(["지역명"])["평당분양가격"].mean()
```

```python
# 전용면적으로 평당분양가격의 평균 구하기
df_last.groupby(["전용면적"])["평당분양가격"].mean()
```

```python
# 지역명, 전용면적으로 평당분양가격의 평균 구하기
df_last.groupby(["지역명", "전용면적"])["평당분양가격"].mean()
```

```python
#unstack()을 이용하면 두번째 인덱스가 컬럼이 됨.
#round()를 이용하면 소수점 생략됨.
df_last.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack().round()
```

```python
# 연도, 지역명으로 평당분양가격의 평균 구하기
# .T/.transpose()를 이용해 행과 열을 바꿈
g = df_last.groupby(["연도", "지역명"])["평당분양가격"].mean()
g
# g.unstack().T
```

## melt로 Tidy data 만들기
* melt를 사용해 열에 있는 데이터를 행으로 녹인다.

```python
# pd.melt를 사용하며 녹인 데이터는 df_first_melt에 담아준다.
df_first_melt = df_first.melt(id_vars="지역", var_name="기간", value_name="평당분양가격")
df_first_melt.head()
```

```python
# df_first_melt의 컬럼 명을 ["지역명", "기간", "평당분양가격"]으로 변경
df_first_melt.columns = ["지역명", "기간", "평당분양가격"]
df_first_melt.head(1)
```

## 연도와 월 분리

```python
date = "2013년12월"
```

```python
date.split("년")
```

```python
date.split("년")[0]
```

```python
date.split("년")[-1].replace("월", "")
```

```python
# '연도'만 반환
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year
parse_year(date)
```

```python
type(parse_year(date))
```

```python
# '월'만 반환
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)
    return month
parse_month(date)
```

```python
# df_first_melt에서 apply를 활용해 '연도'만 추출해서 새로운 컬럼에 담음.
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt.head(1)
```

```python
# df_first_melt에서 apply를 활용해 '월'만 추출해서 새로운 컬럼에 담음.
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```

```python
df_first_prepare = df_first_melt[cols].copy()
df_first_prepare.head(1)
```
